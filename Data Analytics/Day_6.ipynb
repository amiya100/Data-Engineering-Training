HISTOGRAM

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.random_sample(5)

plt.hist(data, color="green", edgecolor = 'yellow', bins=5)

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Plot Example')

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

#Create scatter plot
plt.scatter(x, y, marker='o', color='blue')

#Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

plt.show()
```

```python
categories = ['Housing', 'Utilities', 'Entertainment', 'Food', 'Transportation']
expenses = [1200, 150, 200, 400, 300]

#create pie chart
plt.figure(figsize = (5, 16))

plt.pie(expenses, labels=categories, autopct = '%2.2f%%')

plt.title('Monthly Expenses by Category')
plt.legend()

plt.show()
```

BOX PLOT 
- summarize distribution of numerical data by displaying quartiles, outliers and potential skewmess. Insights into data spreadm central tendency, and variability. Box plots are especially useful for identifying outliers and comparing distributions.

```python
data = [12, 15, 10, 15, 17, 18, 20, 22, 23, 25, 27, 30, 35, 100, -34]

plt.boxplot(data)

plt.title('Boxplot Example')
plt.xlabel('Data')
plt.ylabel('Values')

plt.show()
```

STACK PLOT
- stacked area plot, displays multiple datasets as layers stacked on top of one another, each layer rerpesenting different category or component of data.
- useful for visualizing how individual components contribute to a whole over a continous time peroiod or categorical domain.

```python
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
electronics = [10000, 12000, 11000, 10500]
clothing = [5000, 6000, 7500, 8000]
home_appliances = [3000, 7500, 8200, 9000]

#create stack plot
plt.figure(figsize = (10, 6)) #set figure size, width and height
plt.stackplot(quarters, electronics, clothing, home_appliances, labels = ['Electronics', 'Clothing', 'Home Appliances'])

plt.xlabel('Quarters')
plt.ylabel('Sales ($)')
plt.title('Product Category Sales over Quarters')
plt.legend()

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

#Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

#3D Scatter Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(x, y, z, c=z, cmap = 'viridis')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

#2*2 subplot
fig, axs = plt.subplots(2, 2, figsize = (10, 8))

#Plot data on each subplot
axs[0, 0].plot(x, y1, 'r') #Top-left subplot
axs[ 0, 0].set_title('Sine Wave')
axs[0, 1].plot(x, y2, 'b') #Top-right subplot
axs[ 0, 1].set_title('Cosine Wave')
axs[1, 0].plot(x, y3, 'g') #Bottom-left subplot
axs[ 1, 0].set_title('Tangent Wave')
axs[1, 1].plot(x, y4, 'm') #Bottom-right subplot
axs[ 1, 1].set_title('Exponential Decay')

plt.tight_layout()
plt.show()
```
