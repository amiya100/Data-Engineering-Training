Continue
continue loop - dont run next lines in the loop. Jump to next iterator

```python
li = [1, 2, 3, 4, 5]
for num in li:
    if num % 2 == 0:
        continue
    print(num)
else:
    print("list over")
```

```python
li = ["get", "post", "put", "patch"]
for i in range(5):
    http = input("Enter http method: ")
    if http not in li:
        print("Invalid http method! Stopping...")
        break
    print(http, "method executed")
```

```python
li1 = {1, 2, 3, 4, 5}
li2 = {5, 6}
print(li1 | li2)
print(li1.union(li2))
```
