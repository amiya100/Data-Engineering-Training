- Write code for hcf of two numbers
- Write code for factorial of two numbers

```python
#1. hcf
def hcf(a, b):
    while b:
        a, b = b, a%b
    return a

print(hcf(3, 4))
```

```python
#2. factorial
def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)
print(fact(5))
```

```python
class member:
    def __init__(self, name, age, position = "Developer"):
        self.name = name
        self.age = age
        self.position = position

    def hi(self, name, age):
        self.name = name
        self.age = age

test = member("Joe", 28, "Officer")
#test2 = member("SAN", 29) -error if default arguments not defined
test2 = member("SAN", 29)

print(test.name)
print(test.age)
print(test.position)
print(test2.name)
print(test2.age)
```

INHERITANCE

```python
class Sports:
    def __init__(self, model, color, cost):
        self.__model = model
        self.__color = color
        self.__cost = cost
        
    def get_model(self):
        return self.__model
    def set_model(self, model):
        self.__model = model

    def get_color(self):
        return self.__color
    def set_color(self, color):
        self.__model = color
        
    def get_cost(self):
        return self.__cost
    def set_cost(self, cost):
        self.__cost = cost

class Shoes(Sports):
    def __init__(self, model, color, cost, brand):
        super().__init__(model, color, cost)
        self.__brand = brand

sh1 = Shoes("Shoes", "Black", "Rs. 4000", "ABC")
print(sh1.get_model())
sh2 = Shoes("Shoes", "Black-red", "Rs.3699","XYZ123")
```

CALCULATOR

```python
import sys
while True:
    try:
        n1 = int(input("Enter first number: "))
        n2 = int(input("Enter second number: "))
        op = input("Enter operation (+, -, *, /): ")
        if op == '+':
            res = n1 + n2
        elif op == '-':
            res = n1 - n2
        elif op == '/':
            res = n1 / n2
        elif op == '*':
            res = n1 * n2
        else:
            raise ValueError("Invalid operation")
        print("{0} + {1} = {2}".format(n1, n2, res))
    except:
        print("Error: ", sys.exc_info())
    finally:
        print("****************************")
        if input("Continue (y/n)?") != 'y':
            break
```
