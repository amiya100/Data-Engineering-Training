PYTHON UNIT TESTING

```python
import unittest

def string_validation(username):
    return isinstance(username, str)

def length_validation(username):
    return len(username) >= 8

def first_char_validation(username):
    return username[0].isupper()

class TestUsername(unittest.TestCase):
    def test_valid_string(self):
        self.assertTrue(string_validation('abcd'))
        print("Valid string test passed")
        
    def test_invalid_string(self):
        self.assertFalse(string_validation({}))
        print("Invalid string test passed")

    
    def test_valid_length(self):
        self.assertTrue(length_validation('abcdefghi'))
        print("Valid string length test passed")

    def test_invalid_length(self):
        self.assertFalse(length_validation('abcd'))   
        print("Invalid string length test passed")

    def test_valid_first_char(self):
        self.assertEqual(first_char_validation('Abcdefghij'), True)
        print("Valid first character test passed")
        
    def test_invalid_first_char(self):
        self.assertFalse(first_char_validation('abcd'))
        print("Invalid first character test passed")
            

unittest.main(argv=[''], verbosity=2, exit=False)
```

```python
import unittest
class Bank:
    def __init__(self, balance):
        if not isinstance(balance, (float, int)) or balance < 0:
            raise TypeError
        self.balance = balance
  
    def deposit(self, amnt):
        if not isinstance(amnt, (float, int)):
            raise TypeError
        elif amnt <= 0:
            return "Amount cannot be negative or zero"
        else:
            self.balance += amnt
            return self.balance

    def withdraw(self, amnt):
        if not isinstance(amnt, (float, int)):
            raise TypeError
        elif amnt <= 0:
            return "Amount cannot be negative or zero"
        else:
            self.balance -= amnt
            return self.balance

    def __str__(self):
        return 'Account balance: ' + str(float(self.balance))

class TestBank(unittest.TestCase):
    def test_valid_initialization(self):
        c1 = Bank(0)
        self.assertEqual(str(c1), 'Account balance: 0.0')
        c2 = Bank(2.0)
        self.assertEqual(str(c2), 'Account balance: 2.0')
        c2 = Bank(100)
        self.assertEqual(str(c2), 'Account balance: 100.0')

    def test_invalid_initialization(self):
        with self.assertRaises(TypeError):
            c1 = Bank('abcd')
        with self.assertRaises(TypeError):
            c1 = Bank(-1)


    def test_valid_deposit(self):
        c1 = Bank(0)
        self.assertEqual(c1.deposit(1000), 1000.0)
        
    def test_invalid_deposit(self):
        c1 = Bank(0)
        with self.assertRaises(TypeError):
            c1.deposit('abcd')
        self.assertEqual(c1.deposit(-1), 'Amount cannot be negative or zero')
        self.assertEqual(c1.deposit(0), 'Amount cannot be negative or zero')       
        

    def test_valid_withdraw(self):
        c1 = Bank(1000)
        self.assertEqual(c1.withdraw(300), 700.0)
        
    def test_invalid_withdraw(self):
        c1 = Bank(1000)
        with self.assertRaises(TypeError):
            c1.withdraw('abcd')
        self.assertEqual(c1.withdraw(-1), 'Amount cannot be negative or zero')
        self.assertEqual(c1.withdraw(0), 'Amount cannot be negative or zero') 
        
unittest.main(argv=[''], verbosity=2, exit=False)
```

ITERATORS
- Iterator is an object that allows you to traverse through all elements of a collection one at a time.
- implements __iter__() and __next__() methods
- raises StopIteration when no more data is available
- Used in loops like for loop

```python
class MyNumbers:
    def __iter__(self):
        self.num = 1
        return self

    def __next__(self):
        if self.num <= 5:
            x = self.num
            self.num += 1
            return x
        else:
            raise StopIteration

mynum = MyNumbers()
myiter = iter(mynum)

for x in myiter:
    print(x)
```

```python
class MyCart:
    def __init__(self):
        self.cart = ['Book', 'Pen', 'Pencil', 'Bag']
        
    def __iter__(self):
        self.ind = 0
        return self

    def __next__(self):
        if self.ind < len(self.cart):
            x = self.ind
            self.ind += 1
            return self.cart[x]
        else:
            raise StopIteration

mycart = MyCart()
myiter = iter(mycart)

for x in myiter:
    print(x)
```

GENERATORS
- Special type of iterator defined using function and 'yield' keyword
- used yield instead of return
- generates value lazily (one at a time)
- more memory efficient than normal functions


```python
def my_generator():
    for i in range(1, 6):
        yield i

gen = my_generator()
for value in gen:
    print(value)
```

```python
def my_cart_gen():
    for val in ['Book', 'Pen', 'Pencil', 'Bag']:
        yield val
        
mycart = my_cart_gen()
for val in mycart:
    print(val)
```

ITERATOR VS GENERATOR
- iterator - requires class with __iter__() and __next__()
- generator - uses function with yield keyboard.
- iterator - more code, manual implementation
- generator - simple and concise implementation
- iterator - slightly more control
- generator - more memory efficient

```python
def my_ids():
    for num in range(10, 100, 10):
        yield 'E' + str(num)
        
myids = my_ids()
for val in myids:
    print(val)
```
