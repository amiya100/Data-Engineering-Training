Copies and Views
- slicing creates views - shares address


```python
a = np.arange(10)
b = a[::2]
b
```

```python
c = a[::3]
c
```

```python
print(id(a))
print(id(b))
print(id(c))
```

```python
print(np.may_share_memory(a, b))
print(np.may_share_memory(a, c))
print(np.may_share_memory(b, c))
```

```python
c[0] = 2
a #changes in c modified a as well
```

```python
a = np.arange(10)
c = a[::2].copy()
c
```

```python
c[0] = 2
a #here, with copy() changes in c didnt modify a
```

```python
print(np.may_share_memory(a, c))
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
product_a = [15, 25, 18, 22, 70, 80]
product_b = [35, 40, 50, 30, 70, 65]

#create line plot
plt.plot(months, product_a, marker='o', color = 'pink', linestyle = '--', label = 'Product A')
plt.plot(months, product_b, marker='s', color = 'purple', linestyle = ':', label = 'Product B')

plt.xlabel("Month")
plt.ylabel("Monthly Revenue (in $1000)")
plt.title("Monthly Revenue Comparison")

plt.legend()

plt.show()
```

```python
df = pd. read_csv('IPL_Matches_2008-2020.csv')
df
```

```python
wins = df.groupby('winner')['winner'].count()
toss_won = df.groupby('toss_winner')['toss_winner'].count()

wins_sorted = wins.sort_values(ascending=False)
toss_won_sorted = toss_won.sort_values(ascending=False)

wins_top = wins_sorted.head(8)
toss_won_top = toss_won_sorted.head(8)

team_names = [tname.split()[0] for tname in wins_top.index]

# Setting up the positions for the bars on the x-axis
x = np.arange(len(team_names))

# Bar width for the side-by-side bars
bar_width = 0.4

# Plotting the bars
plt.bar(x - bar_width/2, wins_top.values, bar_width, label='Matches won', color='brown')
plt.bar(x + bar_width/2, toss_won_top.values, bar_width, label='Toss Won', color='orange')

# Adding labels, title, and legend
plt.xticks(x, team_names)
plt.ylabel('Values')
plt.title('Matches Won vs Toss Won')
plt.legend()

# Display the plot
plt.show()
```

```python
city = df.groupby('city')['city'].count().head(6)
plt.plot(city.index, city.values, marker = 'o', markersize=10, color='purple')

plt.xlabel("City")
plt.ylabel("Number of matches played")
plt.title("Number of matches played in each city")

plt.show()
```
