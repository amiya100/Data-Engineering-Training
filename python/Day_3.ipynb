```python
#read from file
f = open("new.txt", "r")
print(f.read())
f.close()
```

```python
#pointer location
f = open("new.txt", "r")
f.seek(5)
print(f.read(6))
print(f.tell())
print(f.read())
print(f.tell())
f.close()
```

```python
#read line by line
f = open("new.txt", "r")
for line in f:
    print(line, end='')
f.close()
```

```python
f = open("new.txt", "r")
print(f.readline(), end='')
print(f.readline())
print(f.readlines())
f.close()
```

```python
f = open("new.txt", "r")
print(f.readlines())
f.close()
```
