DAY 2
1. Why we cant call local variable from outside the function ?
2. Where do we use map, filter and reduce?

ANSWERS
1. Because it has local scope. It exists only while the function runs and gets destroyed after function finishes
2. map() - to apply change to every element (convert student marks to grades, convert prices from usd to inr)
    filter() - keep only items that meet a condition (filter students who passed, get active users from db)
    reduce() - combine elements to single result (total cart price, total salary of employees)
